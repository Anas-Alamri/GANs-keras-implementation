```python
#memory isn't enough to load the enitre dataset :( 
#we will load a subset of it then train on it then reload the subset

def load_training_data_subset(batch_size = 32, no_batches_in_subset = 5):
    '''
    used to load a subset of the dataset to train on, you should call this function every no_batches_in_subset iterations
    '''
    no_of_elements = int(batch_size*no_batches_in_subset)
    #defining the data holding tensors
    A_data = np.zeros((no_of_elements, *A_image_shape))
    B_data = np.zeros((no_of_elements, *B_image_shape))
    #getting random indexes of data paths
    indxs = np.random.randint(0, high=len(train_data_paths), size=no_of_elements)
    #saving the images in the holding tensors
    for i in range(no_of_elements):
        img = cv2.imread(train_data_paths[indxs[i]])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #making it RGB
        shape = img.shape
        #splitting the image into A and B
        A = img[:,:(shape[1]//2),:]  
        B = img[:,(shape[1]//2):,:]
        #saving the A and B images in holding tensors
        A_data[i,:,:,:] = A
        B_data[i,:,:,:] = B
    #normalizing the data
    A_data = A_data/255.0
    B_data = B_data/255.0
    #shuffling the data
    A_data = shuffle(A_data)
    B_data = shuffle(B_data)
    #returning holding tensors
    return A_data, B_data

def load_batch(A_data, B_data, iteration, batch_size = 32, no_batches_in_subset = 5):
    '''
    used to get a data batch out of the data subset
    '''
    start_indx = int(iteration%no_batches_in_subset)
    A_batch = A_data[start_indx:start_indx+batch_size, :,:]
    B_batch = B_data[start_indx:start_indx+batch_size, :,:]
    return A_batch, B_batch
```

```python
G_AB = Sequential()
#encoder
# G_AB.add(ZeroPadding2D(padding=0, input_shape=(256, 256, 3)))
G_AB.add(Conv2D(filters=64, kernel_size=4, strides=2, padding='valid', use_bias=False, input_shape=(256, 256, 3) ))
G_AB.add(LeakyReLU(alpha=0.2))

# G_AB.add(ZeroPadding2D(padding=1))
G_AB.add(Conv2D(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False ))
G_AB.add(BatchNormalization(momentum=0.1))
G_AB.add(LeakyReLU(alpha=0.2))

# G_AB.add(ZeroPadding2D(padding=1))
G_AB.add(Conv2D(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False ))
G_AB.add(BatchNormalization(momentum=0.1))
G_AB.add(LeakyReLU(alpha=0.2))

# G_AB.add(ZeroPadding2D(padding=1))
G_AB.add(Conv2D(filters=64*8, kernel_size=4, strides=2, padding='valid', use_bias=False ))
G_AB.add(BatchNormalization(momentum=0.1))
G_AB.add(LeakyReLU(alpha=0.2))


#decoder

# G_AB.add(ZeroPadding2D(padding=1))
G_AB.add(Conv2DTranspose(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False ))
G_AB.add(BatchNormalization(momentum=0.1))
G_AB.add(Activation('relu'))

# G_AB.add(ZeroPadding2D(padding=1))
G_AB.add(Conv2DTranspose(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False ))
G_AB.add(BatchNormalization(momentum=0.1))
G_AB.add(Activation('relu'))

# G_AB.add(ZeroPadding2D(padding=1))
G_AB.add(Conv2DTranspose(filters=64*1, kernel_size=4, strides=2, padding='valid', use_bias=False ))
G_AB.add(BatchNormalization(momentum=0.1))
G_AB.add(Activation('relu'))

G_AB.add(ZeroPadding2D(padding=1))
G_AB.add(Conv2DTranspose(filters=3, kernel_size=4, strides=2, padding='same', use_bias=False ))
G_AB.add(Activation('sigmoid'))


G_AB.summary()
```

```python
# G_BA = Sequential()
#encoder
# G_BA.add(Conv2D(filters=64, kernel_size=4, strides=2, padding='valid', use_bias=False, input_shape=(256, 256, 3) ))
# G_BA.add(LeakyReLU(alpha=0.2))

# G_BA.add(Conv2D(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False ))
# G_BA.add(BatchNormalization(momentum=0.1))
# G_BA.add(LeakyReLU(alpha=0.2))

# G_BA.add(Conv2D(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False ))
# G_BA.add(BatchNormalization(momentum=0.1))
# G_BA.add(LeakyReLU(alpha=0.2))

# G_BA.add(Conv2D(filters=64*8, kernel_size=4, strides=2, padding='valid', use_bias=False ))
# G_BA.add(BatchNormalization(momentum=0.1))
# G_BA.add(LeakyReLU(alpha=0.2))

# #decoder

# G_BA.add(Conv2DTranspose(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False ))
# G_BA.add(BatchNormalization(momentum=0.1))
# G_BA.add(Activation('relu'))

# G_BA.add(Conv2DTranspose(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False ))
# G_BA.add(BatchNormalization(momentum=0.1))
# G_BA.add(Activation('relu'))

# G_BA.add(Conv2DTranspose(filters=64*1, kernel_size=4, strides=2, padding='valid', use_bias=False ))
# G_BA.add(BatchNormalization(momentum=0.1))
# G_BA.add(Activation('relu'))

# G_BA.add(ZeroPadding2D(padding=1))
# G_BA.add(Conv2DTranspose(filters=3, kernel_size=4, strides=2, padding='same', use_bias=False ))
# G_BA.add(Activation('sigmoid'))


ip = Input(shape=(256, 256, 3))
x = Conv2D(filters=64, kernel_size=4, strides=2, padding='valid', use_bias=False)(ip)
x = LeakyReLU(alpha=0.2)(x)

x = Conv2D(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False )(x)
x = BatchNormalization(momentum=0.1)(x)
x = LeakyReLU(alpha=0.2)(x)

x = Conv2D(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False )(x)
x = BatchNormalization(momentum=0.1)(x)
x = LeakyReLU(alpha=0.2)(x)

x = Conv2D(filters=64*8, kernel_size=4, strides=2, padding='valid', use_bias=False )(x)
x = BatchNormalization(momentum=0.1)(x)
x = LeakyReLU(alpha=0.2)(x)

#decoder

x = Conv2DTranspose(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False )(x)
x = BatchNormalization(momentum=0.1)(x)
x = Activation('relu')(x)

x = Conv2DTranspose(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False )(x)
x = BatchNormalization(momentum=0.1)(x)
x = Activation('relu')(x)

x = Conv2DTranspose(filters=64*1, kernel_size=4, strides=2, padding='valid', use_bias=False )(x)
x = BatchNormalization(momentum=0.1)(x)
x = Activation('relu')(x)

x = ZeroPadding2D(padding=1)(x)
x = Conv2DTranspose(filters=3, kernel_size=4, strides=2, padding='same', use_bias=False )(x)
op = Activation('sigmoid')(x)
G_BA = Model(inputs=ip, outputs=op)

G_BA.summary()
```

```python
D_A = Sequential()

# D_A.add(ZeroPadding2D(padding=0, input_shape=(256, 256, 3)))
D_A.add(Conv2D(filters=64, kernel_size=4, strides=2, padding='valid', use_bias=False, input_shape=(256, 256, 3) ))
D_A.add(LeakyReLU(alpha=0.2))

# D_A.add(ZeroPadding2D(padding=1))
D_A.add(Conv2D(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_A.add(BatchNormalization(momentum=0.1))
D_A.add(LeakyReLU(alpha=0.2))

# D_A.add(ZeroPadding2D(padding=1))
D_A.add(Conv2D(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_A.add(BatchNormalization(momentum=0.1))
D_A.add(LeakyReLU(alpha=0.2))

# D_A.add(ZeroPadding2D(padding=1))
D_A.add(Conv2D(filters=64*8, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_A.add(BatchNormalization(momentum=0.1))
D_A.add(LeakyReLU(alpha=0.2))

# D_A.add(ZeroPadding2D(padding=1))
D_A.add(Conv2D(filters=1, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_A.add(GlobalAveragePooling2D())
D_A.add(Activation('sigmoid'))

D_A.summary()
```

```python
D_B = Sequential()

# D_B.add(ZeroPadding2D(padding=0, input_shape=(256, 256, 3)))
D_B.add(Conv2D(filters=64, kernel_size=4, strides=2, padding='valid', use_bias=False,input_shape=(256, 256, 3) ))
D_B.add(LeakyReLU(alpha=0.2))

# D_B.add(ZeroPadding2D(padding=1))
D_B.add(Conv2D(filters=64*2, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_B.add(BatchNormalization(momentum=0.1))
D_B.add(LeakyReLU(alpha=0.2))

# D_B.add(ZeroPadding2D(padding=1))
D_B.add(Conv2D(filters=64*4, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_B.add(BatchNormalization(momentum=0.1))
D_B.add(LeakyReLU(alpha=0.2))

# D_B.add(ZeroPadding2D(padding=1))
D_B.add(Conv2D(filters=64*8, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_B.add(BatchNormalization(momentum=0.1))
D_B.add(LeakyReLU(alpha=0.2))

# D_B.add(ZeroPadding2D(padding=1))
D_B.add(Conv2D(filters=1, kernel_size=4, strides=2, padding='valid', use_bias=False ))
D_B.add(GlobalAveragePooling2D())
D_B.add(Activation('sigmoid'))

D_B.summary()
```

## Training

```python
#a function used to plot and save the output of the generators, to make us check the outputs of the network
def plot_images(save2file=False, samples=3, step=0):
    filename = "disco_%d.png" % step
    A_batch, B_batch= load_batch(A_data, B_data, iteration = step, batch_size = 3, no_batches_in_subset = 5)
    images_B = (G_AB.predict(A_batch)*255).astype("uint8")
    images_A = (G_BA.predict(B_batch)*255).astype("uint8")
    A_batch = ((A_batch)*255).astype("uint8")
    B_batch = ((B_batch)*255).astype("uint8")
    l_images = [A_batch, images_B, B_batch,images_A]
    f, axrr = plt.subplots(4,samples, figsize=(11,11))
#     print(axrr.shape)
    for i in range(axrr.shape[0]):
        for j in range(axrr.shape[1]):
            axrr[i,j].imshow(l_images[i][j,:,:,:])
            axrr[i,j].axis('off')
    f.tight_layout()
    if save2file:
        f.savefig(filename)
        plt.close('all')
    else:
        plt.show()
```

```python
#some hyper parameters
save_interval = 50   #save a plot that includes the output of the generator every number of steps 
train_steps=2000      # end number of steps
start_step = 0        # start number of steps
batch_size=16        # batch size
no_batches_in_subset = 5   #no of batches in every subset
```

```python
# pretraining the descriminators

A_batch, B_batch= load_batch(A_data, B_data, iteration = 0, batch_size = batch_size, no_batches_in_subset = no_batches_in_subset)
images_fake_A = G_BA.predict(B_batch)   #generating fake images from the generator
images_fake_B = G_AB.predict(A_batch)   #generating fake images from the generator
x_A = np.concatenate((A_batch, images_fake_A))  #stacking both real and fake images
x_B = np.concatenate((B_batch, images_fake_B))  #stacking both real and fake images
#generating labels in which real=1, fake =0
y_A = np.ones([2*batch_size, 1])
y_A[batch_size:, :] = 0
y_B = np.ones([2*batch_size, 1])
y_B[batch_size:, :] = 0

x_A,y_A = shuffle(x_A,y_A)  #shuffling the data and labels
x_B,y_B = shuffle(x_B,y_B)  #shuffling the data and labels

D_A.trainable = True  #making sure that discriminator is trainable
D_B.trainable = True  #making sure that discriminator is trainable
print("traing D_A")
d_A_loss = D_A.fit(x_A, y_A, batch_size=batch_size, epochs=5, validation_split=0.15) #the pretraining step
print("traing D_B")
d_B_loss = D_B.fit(x_B, y_B, batch_size=batch_size, epochs=5, validation_split=0.15) #the pretraining step
```

```python
#the training loop    
for i in range(start_step,train_steps):
    #loading the data 
    if not(i%no_batches_in_subset):
        A_data, B_data = load_training_data_subset(batch_size = batch_size, no_batches_in_subset = no_batches_in_subset)
    A_batch, B_batch= load_batch(A_data, B_data, iteration = i, batch_size = batch_size, no_batches_in_subset = no_batches_in_subset)
   
    ##training the reconstructors
    recon_a_loss = recon_A.train_on_batch(A_batch,A_batch)
#     recon_b_loss = recon_B.train_on_batch(B_batch,B_batch)  #sadly not working :(

    ## training the discriminators
    images_fake_A = G_BA.predict(B_batch)   #generating fake images from the generator
    images_fake_B = G_AB.predict(A_batch)   #generating fake images from the generator
    x_A = np.concatenate((A_batch, images_fake_A))  #stacking both real and fake images
    x_B = np.concatenate((B_batch, images_fake_B))  #stacking both real and fake images
    #generating labels in which real=1, fake =0
    y_A = np.ones([2*batch_size, 1])
    y_A[batch_size:, :] = 0
    y_B = np.ones([2*batch_size, 1])
    y_B[batch_size:, :] = 0
    #shuffling the data and labels
    x_A,y_A = shuffle(x_A,y_A)  
    x_B,y_B = shuffle(x_B,y_B)  
    #making sure that discriminator is trainable
    D_A.trainable = True  
    D_B.trainable = True  
    # the training part
    d_A_loss = D_A.train_on_batch(x_A, y_A)
    d_B_loss = D_B.train_on_batch(x_B, y_B)
    
    ##training the adverserial models
    A_batch, B_batch= load_batch(A_data, B_data, 
                                 iteration = np.random.randint(no_batches_in_subset+1), #made it random to train adverserial on diffrent data than the recon and discrimintator
                                 batch_size = batch_size, 
                                 no_batches_in_subset = no_batches_in_subset) 
    y = np.ones([batch_size, 1]) #the labels, they are always one
    #freeze the discirminators
    D_A.trainable = False 
    D_B.trainable = False 
    #train the adeveserial models, with the discrimimator frozen only the generator is updated\
    # and with all the labels equal (1), we are telling the generator to strive to generate data that will make the]
    # discriminator always output (1) aka fooling it.
    AM_A_loss = AM_A.train_on_batch(B_batch, y) 
    AM_B_loss = AM_B.train_on_batch(A_batch, y) 
    #print the losses and accuracy values 
    log_mesg = "%d: [D_A loss: %f, acc: %f]" % (i, d_A_loss[0], d_A_loss[1])
    log_mesg = "%s  [D_B loss: %f, acc: %f]" % (log_mesg, d_B_loss[0], d_B_loss[1])
    log_mesg = "%s  [recon_A loss: %f, acc: %f]" % (log_mesg, recon_a_loss[0], recon_a_loss[1])
#     log_mesg = "%s  [recon_B loss: %f, acc: %f]" % (log_mesg, recon_b_loss[0], recon_b_loss[1])
    log_mesg = "%s  [AM_A loss: %f, acc: %f]" % (log_mesg, AM_A_loss[0], AM_A_loss[1])
    log_mesg = "%s  [AM_B loss: %f, acc: %f]" % (log_mesg, AM_B_loss[0], AM_B_loss[1])
    print(log_mesg)
    #save the plots of generator outputs
    if save_interval>0:
        if (i+1)%save_interval==0:
            plot_images(save2file=True, samples=3, step=i+1)
```
